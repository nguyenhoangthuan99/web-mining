```python
users = pd.read_csv("ml-100k/u.user",sep="|") 
users.columns
```

```python
users.drop(" zip code",axis=1,inplace=True)
```

```python
users_id = users["user id "]
```

```python
list_occ = ["administrator","artist","doctor","educator","engineer","entertainment","executive","healthcare","homemaker","lawyer","librarian","marketing","none","other","programmer","retired","salesman","scientist","student","technician","writer"]
```

```python
n = len(users_id)
data_u = np.zeros((n,2+len(list_occ)))
```

```python
for index, row in users.iterrows():
    data_u[index][0] = float(row[" age "])/5.
    if row[" gender "] == "M":
        data_u[index][1] = 0
    elif row[" gender "] == "F":
        data_u[index][1] = 2
    else:
        data_u[index][1] = 1
    vec = np.zeros((len(list_occ)))
    idx = list_occ.index(row[" occupation "])
    vec[idx] = 1
    data_u[index][2:] = vec
    
```

```python
data_u
```

```python
from sklearn.cluster import KMeans
clusters = [i*10 for i in range(1,20)]
err = []
for c in clusters:
    kmeans = KMeans(n_clusters=int(c), random_state=0).fit(data_u)
    err.append(kmeans.inertia_)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(clusters,err)
```

```python
n = 100
kmeans = KMeans(n_clusters=n, random_state=0).fit(data_u)
```

```python
user_cluster = {}
for i in range(data_u.shape[0]):
    user_cluster[str(users_id[i])] = int(kmeans.predict([data_u[i]])[0])
```

```python
user_cluster
```

```python
with open("user_cluster_100k.json","w")as fp:
    json.dump(user_cluster,fp)
```

```python
user_neighbor = defaultdict(list)
k=10
for i in range(data_u.shape[0]):
    a = data_u - data_u[i]
    a=np.sum(a**2,axis=-1)
    #print(a)
    user_neighbor[str(users_id[i])] = [int(x) for x in list(a.argsort()[:k]+1)]
    #print(movie_neighbor)
    
```

```python
with open("user_neighbor_100k.json","w")as fp:
    json.dump(user_neighbor,fp)
```

```python
matrix = np.zeros((944,1683))
js = json.load(open("100k/train_100k.json","r"))
for i in range(len(js["users"])):
    matrix[int(js["users"][i]),int(js["items"][i])] = js["rating"][i]
```

```python
matrix
```

```python
user_neighbor = defaultdict(list)
k=11
print(np.sum(matrix[6]>0))
for i in range(1,944):
    list_nei = []
    for j in range(1,944):
        if i == j : continue
        a = matrix[j] - matrix[i]
        b = (matrix[i]>0) == (matrix[j]>0)
        if np.sum(b) == 0: continue
        a = a*b
        a=np.sum(a**2)+5/np.sum(b)
        if a <1:
            list_nei.append((j,a))
    list_nei.sort(key=lambda x :x[1])
    print(i,len(list_nei))
    user_neighbor[str(i)] = [int(x[0]) for x in list_nei[:30]]
with open("user_neighbor_100k_2.json","w")as fp:
    json.dump(user_neighbor,fp)
```
